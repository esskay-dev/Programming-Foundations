```python
# Boolean Expressions: This is used to control the flow of our codes and to build a logic in our conditional statements

# Example:

print(True)
print(False)
print((type)(True))
print(bool(123))
print(bool(("Hello")))
age = 0
print(bool(age))

email = "esskay@gmail.com"
phone_number = '09087654321'
username = ""

# Allows registration
# If any field is filled

print(any([email, phone_number, username])) # True becaue we have at least one information
print(all([email, phone_number, username])) # False because not all the list has information

# isinstance() - to check whether a value belongs to certain data type

print(isinstance(True, bool))
print(isinstance("ess", int))

# endswith and startwith: checks if the strings end or start with a specific word

print("Hello".endswith('o'))
print('hello'.startswith('4'))
```

```python
# Comparison Operators: to compare two values and return True or False based on the result

# Comparison Operators:

# == Equal to
# != Not Equal
# < Less than
# <= Less than or Equal
# > Greater than
# >= Greater than or Equal

# Examples

print(3 > 2) # Value

x_variable = 5

print(x_variable < 2) # Variable

print(2-1 != 2) # Expression

print(len('Esskay') == 3)

print(10 == 10)        # True because 10 is equal to 10
print(10 != 10)        # False because 10 is equal to 10 but the expression says it is not
print(7 > 3)           # True because 7 is greater than 3
print(10 >= 3)         # True because 10 is either greater than 3
print(3 <= 7)          # True because 3 is less than 7


## Note: You can compare strings too alphabetically, not just numbers

# Examples

print("a" < "b")        # True since Python treats strings alphabetically
print("a" == "b")       # False
print('a' == "A")       # False since Python is case sensitive

# Chain Comparison: To check multiple conditions in one line, just like in math (and it evaluates it from left to right, checking each condition one by one)

#example

print(1 < 4 < 6) # True

# practical examples: Is age between 18 and 30?

age = 20

print(18 <= age <=30)
```

```python
# Logical Operators (and | or): Used to combine multiples boolean expressions

print(3 > 1 and 5 < 1)
print(3 < 1 or 5 < 1)

print(3 > 1 or 5 < 1)
print(3 < 1 or 5 < 1)

#checks if the system is under pressure

cpu_usage = 70
memory_usage = 50

print(cpu_usage > 90 or memory_usage > 90)

#checking user credentials before login
email = True
password = True

print(email and password)

#Logican Operators (NOT) - flips the original results

print(not 3 > 2) # False because 3 is greater than 2
print(not 3 < 2) # True
print(not True)
print(not False)

name = ""
print(not name)
print(not 0)

# Execution Order: "and" has higher priority than "or" which means and will be excuted first
# Parenthese () first to control order

# Control Mixed Conditions

# Example
# Allow access only if the user is logged in
# or they are a guest
# but they must not be banned

is_logged_in = True
is_guest = False
is_banned = False

print((is_logged_in or is_guest) and not is_banned)
```

```python
from typing_extensions import final
# @title If Condition

# Conditional Statements = Use to check a condition; if is True? runs the code, if it is Falso, skip it

# If statement: Defines the first condtion; if this is true, do this. otherwise, do nothing

# Standalone: If Statement: Is where everything starts; we use it to define :if this is true, do this; otherwise, do nothing"

# If Rules: Only one if; Starts with if; Required; Can standalone

# Examples

score = 100
if score >= 90:
    print("A")
    print("Great Job!")

# Two-Way Decision: Else Statement

# Runs only if all previous conditions are false

# if nothing was true, do this instead

# Features: comes at the end; no conditions; optional; cannot standalone; only one else

grade = 50
if score >= 100:
  print("AB")
else:
  print("F")

average_score = 90
if average_score >= 100:
  print("Pass Mark")
else:
  print("Fail")


# Multiple COnditions: If-elif-else

# Use to ask a follow-up question
# Only runs if previous conditions were false
# "If the first wasn't true, try this one"

# Rules: comes after if; multiple elif; needs condition; optional; cannot standalone

graded_score = 50
if graded_score >= 90:
  print("A")
elif graded_score >= 80:
  print("B")
else:
  print("Fail")

# multiple elif

student_grade = 35
if student_grade >= 90:
  print("A")
elif student_grade >= 80:
  print("B")
elif student_grade >= 70:
  print("C")
elif student_grade >= 60:
  print("D")
elif student_grade >= 50:
  print("E")
else:
  print("F")

# Nested if: if statement inside another if

Final_grade = 95
submitted_project = True
if Final_grade >= 90:
  if submitted_project:
    print("A+")
  else:
    print("A")
elif Final_grade >= 80:
  print("B")
elif Final_grade >= 70:
  print("C")
elif Final_grade >= 60:
  print("D")
elif Final_grade >= 50:
  print("E")
else:
  print("F")

# Logical Operators (and/or)

Final_result = 55
submitted_project = True
if Final_result >= 90 and submitted_project:
  print("A+")
elif Final_result >= 90:
  print("A")
elif Final_result >= 80:
  print("B")
elif Final_result >= 70:
  print("C")
elif Final_result >= 60:
  print("D")
elif Final_result >= 50 or submitted_project:
  print("E")
else:
  print("F")

# Independent If-else statement
# Each If is checked separately; all conditions are tested, even if one is already true

independent_grade = 50
project = True

if independent_grade >= 90:
  print("High Score")
else:
  print("Low Score")

if project:
  print("Project Submitted")
else:
  print("Project not Submitted")

```

```python
#@title 1 Python Challenge

# Email must not be empty
# Email must contain a "." and "@"
# Email must contain exactly one "@" symbol
# Email must end with ".com", ".org", or ".net"
# Email must not be longer than 254 characters
# Email must start and end with a letter or digit

email = "salako@gmail.com"
valid = True

# clean the String
email = email.strip()

# Email must not be empty
if email == "":
  print("Email cannot not be empty.")
  valid = False

# Email must contain a "." and "@"
if not ("." in email and "@" in email):
  print("Email must contain . and @")
  valid = False

# Email must contain exactly one "@" symbol
if (email.count("@") != 1):
  print('Email must contain exactly one "@" symbol')
  valid = False

# Email must end with ".com", ".org", or ".net"
if not email.endswith(("com", "org", "net")):
  print("Email must end with .com, .org, or .net")
  valid = False

# Email must not be longer than 254 characters
if len(email) > 254:
  print("Email must not be longer than 254 characters")
  valid = False

# Email must start and end with a letter or digit
if not (email[0].isalnum() and email[-1].isalnum()):
  print("Email must start and end with a letter or digit")
  valid = False

if valid:
  print("Email is valid")
```

```python
# @title 2 Python Challenge

# Validate the quality and correctness of Passwords

# Must not be empty
# Must be at least 8 characters
# Must include at least 1 uppercase
# Must include at least 1 lowercase
# Must not be same as the email
# Must not contain any spaces
# Must start and end with a letter or digit

password = "SALAKO 886744"
# Assuming 'email' is defined from previous cells or kernel state.
# For testing purposes, let's ensure it's available.
email = "salako@gmail.com"

valid_password = True # Initialize a boolean to track validity

# 1. Must not contain any spaces (fixed by removal at the start)
password = password.replace(' ', '')

# Validation Checks
# 1. Must not be empty
if not password:
    print("Password cannot be empty")
    valid_password = False
# 2. Must be at least 8 characters
elif len(password) < 8:
    print("Password must be at least 8 characters")
    valid_password = False
# 3. Must include at least 1 uppercase
elif not any(char.isupper() for char in password):
    print("Password must include at least 1 uppercase")
    valid_password = False
# 4. Must include at least 1 lowercase
elif not any(char.islower() for char in password):
    print("Password must include at least 1 lowercase")
    valid_password = False
# 5. Must not be same as the email
elif password == email:
    print("Password must not be the same as email")
    valid_password = False
# 6. Must not contain any spaces (this check is effectively covered by the initial replace, but for explicit validation flow, we could add it)
#    However, since we removed spaces, this condition would always pass if placed here.
# 7. Must start and end with a letter or digit
elif not (password[0].isalnum() and password[-1].isalnum()):
    print("Password must start and end with a letter or digit")
    valid_password = False

# Final check
if valid_password:
    print("Password is valid")
```

```python
#@title Match Case Statement

# Evaluate a value against multiple values
# Runs the code of the first match

# Example: Convert the full country names into 2-letter abbrevations

country ="United Statets"

match country:
  case "United States":
    print("US")
  case "Nigeria":
    print("NG")
  case "Ghana":
    print("GH")
  case _:
    print("Unknown COuntry")

```
